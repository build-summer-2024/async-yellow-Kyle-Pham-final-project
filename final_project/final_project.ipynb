```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
 
nfl_data = pd.read_csv("team_stats_2003_2023.csv")
nfl_data.iloc[576:671]
```

```python
nfl_data.info()
nfl_data.describe()
```

```python
top_teams = nfl_data.groupby('team')['wins'].sum().nlargest(5).index

top_nfl = nfl_data[nfl_data['team'].isin(top_teams)]

# Grafics
top_nfl.groupby(['year', 'team']).sum()['wins'].unstack().plot(kind='line', figsize=(12, 6))
plt.title('Wins per Team Over the Years (Top 5 Teams)')
plt.xlabel('Year')
plt.ylabel('Wins')
plt.legend(loc='upper right')
plt.show()
```

```python
plt.figure(figsize=(12,6))
top_10_points = nfl_data.groupby("team")["points"].sum().nlargest(10)
nfl_data_top_10_points = nfl_data[nfl_data["team"].isin(top_10_points.index)]
sns.barplot(top_10_points, palette="deep")
plt.title("Top 10 Teams With Total Highest Points")
plt.xlabel("Teams")
plt.ylabel("Points")
plt.xticks(rotation=45)
plt.show()
```
