```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt 
 
nfl_data = pd.read_csv("team_stats_2003_2023.csv")
nfl_data.iloc[576:671]
```

```python
nfl_data.info()
nfl_data.describe()
```

```python
## Question 1

top_teams = nfl_data.groupby('team')['wins'].sum().nlargest(5).index

top_nfl = nfl_data[nfl_data['team'].isin(top_teams)]

# Grafics
top_nfl.groupby(['year', 'team']).sum()['wins'].unstack().plot(kind='line', figsize=(12, 6))
plt.title('Top 5 Teams Wins Over the Years')
plt.xlabel('Year')
plt.ylabel('Wins')
plt.legend(loc='upper right')
plt.show()
```

```python
## Question 2

plt.figure(figsize=(12,6))
top_10_points = nfl_data.groupby("team")["points"].sum().nlargest(10)
nfl_data_top_10_points = nfl_data[nfl_data["team"].isin(top_10_points.index)]
sb.barplot(top_10_points)
plt.title("Top 10 Teams With Total Highest Points")
plt.xlabel("Teams")
plt.ylabel("Points")
plt.xticks(rotation=45)
plt.show()
```

```python
## Question 3

top_5_rush_tds =  nfl_data.groupby("team")["rush_td"].sum().nlargest(5)
palette_color = sb.color_palette("bright")
plt.pie(top_5_rush_tds, labels=top_5_rush_tds.index, autopct="%1.1f%%", colors=palette_color)
plt.title("Top 5 Teams with Most Rushing TDs'")
plt.show()
```

```python
## Question 4
plt.figure(figsize=(12,6))
top_5_penalties = nfl_data.groupby("team")["penalties"].sum().nlargest(5)
nfl_top_5_penalites = nfl_data[nfl_data['team'].isin(top_5_penalties.index)]
sb.scatterplot(top_5_penalties)
plt.title("Top 5 Teams with Penalties (2003-2023)")
plt.xlabel("Teams")
plt.ylabel("Penalties")
plt.xticks(rotation=45)
plt.show()
```

```python
## Question 5: Which 5 teams had the most points scored on them (worst defense)?

plt.figure(figsize=(12,6))
top_5_worst_defense = nfl_data[nfl_data["year"] == 2023].groupby("team")["points_opp"].sum().nlargest(5)
palette_color = sb.color_palette("bright")
plt.barh(top_5_worst_defense.index, top_5_worst_defense, color=palette_color)
plt.xlabel('Total Points Allowed')
plt.ylabel('Team')
plt.title('Top 5 Worst Defenses by Points Allowed in 2023')
plt.show()
```

```python
## Question 6: Who were the 5 teams to have the highest winning percentage overall?

plt.figure(figsize=(12,6))
top_5_winning_per = nfl_data[nfl_data["year"] == 2023].groupby("team")["win_loss_perc"].sum().nlargest(5)
palette_color = sb.color_palette("bright")
plt.plot(top_5_winning_per.index, top_5_winning_per, marker='o', color='red')
plt.xlabel('Teams')
plt.ylabel('Winning Percentage')
plt.title('Top 5 Teams with the Highest Winning Percentage in 2023')
plt.grid(True)
plt.show()
```

```python
## Question 7: Which 5 teams had the worst winning percentage in 2022?

plt.figure(figsize=(12,6))
top_5_losing_per = nfl_data[nfl_data["year"] == 2022].groupby("team")["win_loss_perc"].sum().nsmallest(5)
palette_color = sb.color_palette("bright")
plt.plot(top_5_losing_per.index, top_5_losing_per, marker='o', color='blue')
plt.xlabel('Teams')
plt.ylabel('Losing Percentage')
plt.title('Top 5 Teams with the Worst Losing Percentage in 2022')
plt.grid(True)
plt.show()
```

```python
## Question 8: Who are the Top 5 teams with the highest pass completion rate (2023)?
filtered_data = nfl_data[nfl_data["year"] == 2023]

team_pass_per = filtered_data.groupby("team")[["pass_cmp", "pass_att"]].sum()

team_pass_per["completion_rate"] = team_pass_per["pass_cmp"] / team_pass_per["pass_att"]

top_5_teams_pass = team_pass_per.nlargest(5, 'completion_rate')

plt.figure(figsize=(12, 6))
sb.lineplot(x='team', y='completion_rate', data=top_5_teams_pass, marker='o')
plt.title("Top 5 Teams with Highest Pass Completion Rate (2023)")
plt.xlabel("Teams")
plt.ylabel("Pass Completion Rate")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
filtered_data = nfl_data[nfl_data["year"] == 2023]

total_att = filtered_data.groupby("team")[["rush_att", "pass_att"]].sum()

total_att["completion_rate"] = total_att["rush_att"] + total_att["pass_att"]

top_5_total_att = total_att.nlargest(5, 'completion_rate')

plt.figure(figsize=(12, 6))
sb.scatterplot(x='team', y='completion_rate', data=top_5_total_att, marker='s', color='red')
plt.title("Top 5 Teams with Highest Offensive Attempts (2023)")
plt.xlabel("Teams")
plt.ylabel("Offensive Attempts")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
## - [ ] Question 10: What are the total yards gained from the top 5 teams in the NFL 2023?

filtered_data = nfl_data[nfl_data["year"] == 2023]

total_yds = filtered_data.groupby("team")[["rush_yds", "pass_yds"]].sum()

total_yds["total_yards"] = total_yds["rush_yds"] + total_yds["pass_yds"]

plt.figure(figsize=(12, 6))
top_5_total_yds = total_yds.nlargest(5, 'total_yards')

plt.figure(figsize=(12, 6))
plt.barh(top_5_total_yds.index, top_5_total_yds['total_yards'], color=sb.color_palette("bright"))
plt.xlabel('Total Yards')
plt.ylabel('Team')
plt.title('Top 5 Teams with Highest Total Yards in 2023')
plt.show()
```
